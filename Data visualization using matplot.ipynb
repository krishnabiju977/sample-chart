```python
import pandas as pd
import matplotlib.pyplot as plt
s=pd.read_csv(r"C:Documents/Salaries modi.csv")
s
```

```python
#LINE CHART
plt.plot(s['salary'])
plt.title('salary information')
plt.xlabel('salary count')
plt.ylabel('salaries in thousand')
plt.show()
```

# BAR CHART

```python
plt.bar(x=s['sex'],height=s['phd'])
plt.show()
```

```python
#HISTOGRAM
```

```python
plt.hist(s['salary'],bins=20,histtype='bar',color='green')
plt.show()
```

```python
#SCATTER PLOT
```

```python
plt.scatter(s['salary'],s['service'],marker='*',c='red')
plt.xlabel('salary')
plt.ylabel('service')
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
#style.use('ggplot')
#style.use('fivethirtyeight')
style.use('dark_background')
plt.plot(s['service'],linewidth=3,c='b')
plt.scatter(s['phd'],s['service'],linewidth=1,marker='*',c='y')
plt.title('Epic Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.grid(True,color='r',linestyle='-',linewidth=3)
plt.show()
```

```python
style.use('grayscale')
x=[5,8,10]
y=[12,16,6]
x2=[5,9,11]
y2=[6,15,3]
plt.plot(x,y,'g',label='line one',linewidth=5,linestyle="-")
plt.plot(x2,y2,'c',label='line two',linewidth=5,linestyle="--")
plt.title('Epic Info')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.grid(True,color='y',linestyle='-',linewidth=3)
plt.show()
```

```python
plt.plot(s['phd'],'y',label='line one',linewidth=3,linestyle='-')
plt.plot(s['service'],'c',label='line two',linewidth=5,linestyle='-.')
plt.title('Epic Info')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
```
